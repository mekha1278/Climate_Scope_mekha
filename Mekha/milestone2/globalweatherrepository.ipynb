```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean(numeric_only=True))
```

```python
df = df.fillna("Unknown")
df.isnull().sum()
```

```python
df['last_updated'] = pd.to_datetime(df['last_updated'], errors='coerce')
df['last_updated'].dtypes
```

```python
df = df.drop(columns=['temperature_fahrenheit', 'feels_like_fahrenheit', 'pressure_in', 'precip_in', 'visibility_miles', 'gust_mph', 'wind_mph'], errors='ignore')
```

```python
df['year'] = df['last_updated'].dt.year
df['month'] = df['last_updated'].dt.month
df['hour'] = df['last_updated'].dt.hour

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['season'] = df['month'].apply(get_season)

df[['last_updated', 'year', 'month', 'hour', 'season']].head()
```

```python
df = df[(df['temperature_celsius'] > -50) & (df['temperature_celsius'] < 60)]
df = df[(df['humidity'] >= 0) & (df['humidity'] <= 100)]
```

```python
df['date'] = df['last_updated'].dt.date
daily_df = df.groupby('date')[['temperature_celsius', 'humidity', 'wind_kph']].mean().reset_index()

daily_df.head()
```

```python
df.to_csv("Cleaned_GlobalWeatherRepository.csv", index=False)

```
